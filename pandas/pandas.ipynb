```python
# Data analysis with python and pandas
# create account in Quandl
import quandl as qd
import pandas as pd
import pickle

api_key = 'v7mGUMf4_SfyhSzMXsWM'
# df = qd.get('FMAC/HPI_AK', authtoken = api_key)
# df.head()
# fiddy_state is a list of data frame

def state_list():
    fiddy_state = pd.read_html('https://simple.wikipedia.org/wiki/List_of_U.S._states')
    return fiddy_state[0][0][1:]


def grab_initial_state_date():
    states = state_list()
    main_df = pd.DataFrame()

    for abbv in states:
        query = "FMAC/HPI_"+str(abbv)    
        df = qd.get(query, authtoken=api_key)
        df.rename(columns={'Value':str(abbv)}, inplace=True)

        if main_df.empty:
            main_df = df
        else:
            main_df = main_df.join(df)

    pickle_out = open('fiddy_states.pickle','wb')
    pickle.dump(main_df, pickle_out)
    pickle_out.close()


grab_initial_state_date()


```
